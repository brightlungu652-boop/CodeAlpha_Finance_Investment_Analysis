```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
portfolio = pd.DataFrame({

    "Asset":[
        "Microsoft",
        "Apple",
        "NVIDIA",
        "VOO ETF",
        "QQQ ETF",
        "Bitcoin",
        "Ethereum",
        "Solana"
    ],

    "Type":[
        "Stock",
        "Stock",
        "Stock",
        "ETF",
        "ETF",
        "Crypto",
        "Crypto",
        "Crypto"
    ],

    "Allocation (%)":[
        25,
        20,
        15,
        15,
        10,
        8,
        5,
        2
    ]

})

portfolio
```

```python
portfolio_value = 100000  # INR

portfolio["Investment (₹)"] = (
    portfolio["Allocation (%)"]/100
) * portfolio_value

portfolio
```

```python
portfolio["Rationale"] = [

    "Stable technology leader with strong cloud business",

    "Diversified consumer technology company",

    "High-growth AI semiconductor company",

    "Broad exposure to the S&P 500",

    "Exposure to leading NASDAQ technology companies",

    "Largest cryptocurrency with relatively lower volatility",

    "Leading smart contract platform",

    "High-growth cryptocurrency with higher risk"

]

portfolio
```

```python
portfolio
```

```python
asset_summary = portfolio.groupby("Type")[
    "Investment (₹)"
].sum()

asset_summary
```

```python
plt.figure(figsize=(7,7))

plt.pie(
    asset_summary,
    labels=asset_summary.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Portfolio Allocation by Asset Class")

plt.tight_layout()

plt.savefig(
    "figures/asset_class_allocation.png",
    dpi=300
)

plt.show()
```

```python
import os

os.makedirs("exports", exist_ok=True)
os.makedirs("figures", exist_ok=True)

portfolio.to_excel(
    "exports/portfolio.xlsx",
    index=False
)

asset_summary.to_excel(
    "exports/portfolio_summary.xlsx"
)
```
